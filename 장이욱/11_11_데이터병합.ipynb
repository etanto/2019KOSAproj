# 원본 데이터 불러오기

```python
df1 = pd.read_csv("hn14_all.csv")
df2 = pd.read_csv("hn15_all.csv")
df3 = pd.read_csv("hn16_all.csv")

df4 = pd.read_csv("hn14_24rc.csv")
df5 = pd.read_csv("hn15_24rc.csv")
df6 = pd.read_csv("hn16_24rc.csv")

df7 = pd.read_csv("hn14_ffq.csv")
df8 = pd.read_csv("hn15_ffq.csv")
df9 = pd.read_csv("hn16_ffq.csv")
```

```python
df4
```

```python
print(df1.shape, df2.shape, df3.shape)
print(df4.shape, df5.shape, df6.shape)
print(df7.shape, df8.shape, df9.shape)
```

```python
ls1 = []
for i in df1.columns:
    if i in df2.columns:
        if i in df3.columns:
            ls1.append(i)
```

```python
ls2 = []
for i in df4.columns:
    if i in df5.columns:
        if i in df6.columns:
            ls2.append(i)
```

```python
ls3 = []
for i in df7.columns:
    if i in df8.columns:
        if i in df9.columns:
            ls3.append(i)
```

```python
#데이터 세로로 병합
df_all_1 = pd.concat([df1[ls1],df2[ls1],df3[ls1]], ignore_index=True)
df_all_2 = pd.concat([df4[ls2],df5[ls2],df6[ls2]], ignore_index=True)
df_all_3 = pd.concat([df7[ls3],df8[ls3],df9[ls3]], ignore_index=True)
```

```python
print(df_all_1.shape, df_all_2.shape, df_all_3.shape)
```

# 저장

```python
df_all_1.to_csv("df_all.csv" ,index=False,encoding='euc_kr')
```

```python
df_all_2.to_csv("df_24rc.csv",index=False,encoding='euc_kr')
```

```python
df_all_3.to_csv("df_ffq.csv" ,index=False,encoding='euc_kr')
```
