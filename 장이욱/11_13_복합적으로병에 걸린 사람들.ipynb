# 데이터 불러오기 data

```python
data = pd.read_csv('mid2_2.csv',encoding='euc-kr')
```

# 질병 별 환자수 데이터 프레임 df1 생성

```python
#질병 별 환자수
#데이터 프레임 컬럼이름만 설정해서 생성
df1 = pd.DataFrame(columns=["name","healthy_count","sick_count",
                            "healthy_rate","sick_rate"] )
#데이터 프레임에 행추가 방식으로 값을 넣어준다.
for i, col in enumerate(data.filter(regex = '^D.{2}_pr').columns):
    df1.loc[i] = [data[col].name] +\
                data[col].value_counts().tolist()+\
                data[col].value_counts(True).tolist()
```

```python
#데이터 프레임에 병 이름을 나타내는 컬럼 추가
tmp = ["고혈압 현재 유병여부","이상지질혈증현재 유병여부",
                  "뇌졸증","심근경색증","협심증","골관절염","류머티스",
                  "폐결핵","천식","당뇨","갑상선","위암","간암",
                  "대장암","유방암","자궁경부암","폐암","갑상선암",
                  "우울증","아토피피부염","알레르기","신부전",
                  "B형간염","C형간염","간경변증"]

df1["name2"] = tmp
```

# 복합 질병 여부와 비율 데이터프레임 각각 

```python
# 함수 : 질병에 따라 복합질병 환자 여부
# 데이터프레임으로 반환

def double1(data=data,df1=df1, x="DE1_pr", q = "q1"):
    #컬럼 이름 생성 (df1 필요)
    tmp = []
    for i in df1.name:
        tmp.append(i + q)
    
    #임시 데이터프레임 생성 
    df3 = pd.DataFrame(columns = tmp)
    
    #요소별 곱을 통해 다른 질병과 같이 걸린 여부 생성
    #(data, df1 필요)
    for i in range(len(df1)):
        tmp2 = data[x] * data[df1.name[i]]
        df3.iloc[:,i] = tmp2
    
    #복합적으로 걸린 사람 비율
    #임시 데이터프레임 생성
    df4 = pd.DataFrame(columns=["name","name2","sum","rate"])
    for i in range(len(df1)):
        df4.loc[i] = [df1.name[i]]+\
                     [df1.name2[i]]+\
                     [df3.sum()[i]]+\
                     [round(df3.sum()[i]/df1.sick_count[i], 3)]
                            
    return df3, df4
```

```python
df3, df4 = double1(data=data,df1=df1, x="DE1_pr", q="q1")
```
