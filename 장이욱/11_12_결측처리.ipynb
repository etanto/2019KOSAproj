```python
df = pd.read_csv("mid.csv",encoding='euc_kr')
```

```python
df_grouped = df.groupby(by=["age","sex"]).mean().reset_index()
```

### 신장, 체중 채우기 HE_ht, HE_wt

```python
df.HE_ht.fillna("nan",inplace=True)
for i in range(len(df)):
    if df.HE_ht[i] == "nan":
        df.HE_ht[i] = float(df_grouped.loc[(
            df_grouped.age == df.age[i])&(
            df_grouped.sex==df.sex[i]),"HE_ht"])
```

```python
df.HE_ht.isnull().sum()
```

```python
df.HE_wt.fillna("nan",inplace=True)
for i in range(len(df)):
    if df.HE_wt[i] == "nan":
        df.HE_wt[i] = float(df_grouped.loc[(
            df_grouped.age == df.age[i])&(
            df_grouped.sex == df.sex[i]),"HE_wt"])
```

```python
df.HE_wt.isnull().sum()
```

### 허리둘레 채우기 HE_wc

```python
df.HE_wc.fillna("nan",inplace=True)
for i in range(len(df)):
    if df.HE_wc[i] == "nan":
        df.HE_wc[i] = float(df_grouped.loc[(
            df_grouped.age == df.age[i])&(
            df_grouped.sex == df.sex[i]),"HE_wc"])
```

```python
#데이터 타입 변경
df.HE_wc = pd.to_numeric(df.HE_wc)
```

```python
df.HE_wc.isnull().sum()
```
