```python
import pandas as pd
```

```python
data = pd.read_csv('./건강/mid2.csv', encoding ='euc-kr')
```

```python
def func(x):
    if x in [8, 9]:
        x = 0
    else:
        pass
    return x
    
```

```python
for col in data.filter(regex = '^D.{2}_pr').columns:
    data[col] = data[col].map(lambda x : func(x))
```

```python
for col in data.filter(regex = '^D.{2}_pr').columns:
    print(data[col].value_counts())
```
