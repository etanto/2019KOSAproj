```python
yang_ls = ['열량','탄수화물','당류','식이섬유','지방','포화지방산',
          '트랜스지방산','콜레스테롤','단백질','수분','루신','이소루신',
          '발린','라이신','트레오닌','히스티딘','비타민A','비타민D',
          '비타민K','비타민C','티아민','리보플라빈','비타민B6','엽산',
          '비타민B12','칼슘','인','나트륨','칼륨','마그네슘','철',
          '아연','구리','망간','요오드','셀레늄','메티오닌','시스테인',
          '페닐알라닌','티로신']
```

```python
def sortyang(strstr,cname):
    check=0
    name = []
    ind = []
    for num, i in enumerate(data.values):
        for j in i:
            if check==2:
                name.append(j)
                ind.append(num)
                check=0;
                break
            if j == str(strstr):
                check=2;
            
    nm_df = pd.DataFrame(name,columns=[str(cname)],index=ind)
    #nm_s = pd.to_numeric(nm_df[str(cname)])
    return nm_df
```

```python
aa = []
for i in  yang_ls:
    try:
        tmp=sortyang(i,i)
        aa.append(tmp)
    except:
        print(i)
    
```

```python
temp = aa[0]
```

```python
len(aa)
```

```python
for i in range(1,len(aa)):
    temp = pd.concat([temp, aa[i]], join='outer', axis=1)
    
```

```python
temp.to_csv('temp.csv',encoding='UTF-8-sig')
```
